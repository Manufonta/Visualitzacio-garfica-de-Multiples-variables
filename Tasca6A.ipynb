### Géneros

```python
movies.groupby('genero').count()
```

Para tratar los géneros, parece que lo más indicado será hacer una data frame dummies

```python
todos_generos = []
for x in movies['genero']:
    todos_generos.extend(x.split('|'))
```

```python
len(todos_generos)
```

```python
generos = pd.unique(todos_generos)
generos
```

```python
ceros = np.zeros((len(movies),len(generos)))
ceros
```

```python
dummies = pd.DataFrame(ceros, columns = generos)
dummies
```

```python
for i, gen in enumerate(movies.genero):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

```python
dummies.head()
```

```python
movies.head()
```

```python
dummies.sum()
```

### Análisis de los géneros

Antes de continiar y hacer un join con el dataset movies, podemos sacar alguna información sobre la distribución de géneros

```python
suma_generos = pd.DataFrame({'genero' : generos,
                            'numero' : dummies.sum()}
                           ).set_index(np.arange(18))
suma_generos
#dummies.sum()
```

```python
suma_generos.describe()
```

```python
ax = sns.boxplot(y = 'numero', data = suma_generos, color = 'lightblue', width = 0.5)
ax = sns.stripplot(y = 'numero', data = suma_generos, color = 'darkblue', alpha = 0.7)
```

Vemos que la gran mayoría de los géneros están repetidos unos pocos cientos de veces, alrededor de 200. 
Sin embargo, tenemos dos outliers, dos géneros que se repiten mucho más que el resto, con 1600 y 1200 películas.

```python
ax = sns.barplot(x = 'genero', y = 'numero', data = suma_generos.sort_values('numero', ascending = False))
plt.xticks(rotation=75)
plt.show()
```

Con este gáfico de barras identificamos que se trata de Drama y Comedy

```python
dummies_coocurrence = dummies.T.dot(dummies)
dummies_coocurrence
```

```python
dummies_cooc = dummies_coocurrence
for i, gen in enumerate(dummies_coocurrence):
    dummies_cooc.iloc[[i]] = dummies_coocurrence.iloc[[i]]/dummies_coocurrence.iloc[i,i]*100

ax = plt.subplots(figsize = (15,12))

sns.heatmap(dummies_cooc, linewidths = .5 , annot = True, fmt=".3g")
```

En este mapa de calor se muestra, para cada fila, qué porcentaje de películas etiquetadas con el género de esa fila tienen también asignado el género de cada columna.

Realizamos un join para juntar toda la información y poder proceder con otros análisis:

```python
movies_dummies = movies.join(dummies)
movies_dummies.drop(columns = ['titulo_or','n_generos'], inplace = True)
movies_dummies
```

```python
palabras_melted = movies_dummies.iloc[:,4:23].melt(id_vars=["palabras"], var_name="genero", value_name="numero")
palabras_melted
```

```python
palabras_melted['numero'].unique()
```

```python
ax = plt.subplots(figsize = (18,10))
ax = sns.violinplot(data = palabras_melted[palabras_melted['numero']==1], x='genero', y='palabras' )

ax.set_title('Palabras en el título por género', size = 20)
ax.set_ylabel('Palabras en el título', size = 25)
ax.set_xlabel(' ')
```

Ahora podemos observar, desplegado por géneros, el número de palabras en el título.
Destacan algunos géneros con una distribución muy alargada (Fantasy, Documentary, Animation...) y otros más concentrados en la zona de 2.5 palabras (Thriller, Mistery, Film-Noir...)

```python
movies_melted = movies_dummies.iloc[:,3:23].melt(id_vars=["año","palabras"], var_name="genero", value_name="numero")
movies_melted.drop(columns = 'palabras', inplace = True)
movies_melted
```

```python
movies_melted_ano = movies_melted.groupby(['genero','año']).sum().reset_index()
movies_melted_ano
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot(111)

ax = sns.lineplot(data = movies_melted_ano, x = 'año', y = 'numero', hue = 'genero')

plt.show()
```

Aunque sería posible sacar alguna conclusión de esta gráfica, en general es muy difícil de ver y distinguir entre géneros. Para ello, probemos mejor a separar las representaciones por génro.

```python
sns.set_theme(style = 'dark')
ax = sns.relplot(
    data = movies_melted_ano,
    x = 'año', y = 'numero', hue = 'genero',
    col = 'genero', col_wrap = 3, kind = 'line',
    palette = 'crest', linewidth = 6,
    zorder = 5, aspect = 1.5, legend = False)

for gen, ax in ax.axes_dict.items():
    ax.text(1920, 144, gen, fontweight="bold", fontsize = 20)
    ax.set_title('')
    ax.set_xlabel('')
    ax.set_ylabel('Número', fontsize = 17)
```

De esta gráfica podemos decir que todos los géneros tienen una distribución parecida con el tiempo, aumentando ligeramente en los 80 y mucho más en los 90.
Una excepción podría ser Horor, cuyo pico está en los 80 y no en los 90.

Separación por décadas:

```python
bins = [1919,1929,1939,1949,1959,1969,1979,1989,2001]
nombres = ['20s','30s','40s','50s','60s','70s','80s','90s']
decadas = pd.cut(movies_melted_ano['año'], bins, labels = nombres, right = False)
decadas
```

Realmente las décadas no son exactas, se ha incluído 1919 en los 20s (no afecta mucho, ya que solo hay una película) y al 2000 en los 90s, para no dejar un año descolgado.

```python
movies_melted_ano['decada'] = decadas
movies_melted_ano
```

```python
movies_melted_decadas = movies_melted_ano.groupby(['genero','decada','año']).agg(sum).reset_index()
movies_melted_decadas
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot(111)
ax = sns.lineplot(data = movies_melted_ano[movies_melted_ano['decada'] == '80s']
                  , x = 'año', y = 'numero', hue = 'genero', palette = 'Paired')
plt.show()
```

```python
mov_decadas = movies_melted_decadas[['genero','decada','numero']].groupby(['decada','genero']).agg(sum).reset_index()
mov_decadas = mov_decadas.pivot(index="genero", columns="decada", values="numero")
mov_decadas = pd.DataFrame(np.array(mov_decadas),columns = ['20s', '30s', '40s', '50s', '60s', '70s', '80s', '90s'])
mov_decadas['genero'] = pd.DataFrame(generos, columns = ['genero']).sort_values('genero').reset_index()['genero']
mov_decadas.sort_values('90s',ascending = False, inplace = True)
mov_decadas
```

```python
prueba = pd.DataFrame(generos, columns = ['genero']).sort_values('genero').reset_index()
prueba
```

```python
sns.set_theme(style = 'darkgrid')

ax = sns.PairGrid(mov_decadas,
                 x_vars=mov_decadas.columns[0:-1], y_vars=['genero'],
                 height=6, aspect=.30)
sns.stripplot
ax.map(sns.stripplot, size=10, orient="h", jitter=False,
      palette="flare_r", linewidth=1, edgecolor="w")

ax.set(xlabel="N peliculas", ylabel="")

titles = ['20s', '30s', '40s', '50s', '60s', '70s', '80s', '90s']

for axx, title in zip(ax.axes.flat, titles):

    axx.set(title=title)
    axx.xaxis.grid(False)
    axx.yaxis.grid(True)
```

Con la gráfica anterior podemos ver la distribución de los géneros de las pelícualas en una década específica.
Vemos variaciones interesantes en la posición de los géneros con más películas, como Musical en los 30s, Film-Noir en los 40s, Sci-Fi en los 50 o Adventure, Horror y Action enre los 60s y los 80s.

```python
totales_ano = movies_melted_ano[['año','numero']].groupby('año').agg(sum)
totales_ano
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot(111)

ax = sns.barplot(data = totales_ano.reset_index(), x = 'año', y = 'numero', palette = 'Paired')

ax.set_xticks(ax.get_xticks()[1:1:5])

plt.show()
```

Con la gráfica anterior, podemos ver de una nueva forma la distribución del número de películas en el tiempo. 
Como ya veníamos viendo, la cantidad de películas anuales en los 90s es hasta 5-10 veces mayor que en años de otras décadas.

```python
movies_melted_ano['total_año'] = movies_melted_ano['año'].map(totales_ano['numero'].to_dict())
movies_melted_ano
```

```python
movies_melted_ano['porcentage'] = movies_melted_ano['numero']/movies_melted_ano['total_año']*100
movies_melted_ano
```

```python
fig = plt.figure(figsize=(14,8))
ax = plt.subplot(111)
ax = sns.lineplot(data = movies_melted_ano
                  , x = 'año', y = 'porcentage', hue = 'genero', palette = 'Paired')
plt.show()
```

Ahora podemos ver una gráfica parecida a la presentada previamente, pero con porcentages en lugar de totales. Con porcentajes es más sencillo distinguir las líneas, aunque sigue siendo una gráfica basante sucia.

```python
sns.set_theme(style = 'dark')
ax = sns.relplot(
    data = movies_melted_ano,
    x = 'año', y = 'porcentage', hue = 'genero',
    col = 'genero', col_wrap = 3, kind = 'line', palette = 'crest',
    linewidth = 6, aspect = 1.5, legend = False)

for gen, ax in ax.axes_dict.items():
    ax.text(1930, 90, gen, fontweight="bold", fontsize = 20)
    ax.set_title('')
    ax.set_xlabel('')
    ax.set_ylabel('Número', fontsize = 14)
```

Es interesante ver que las fluctuaciones en la primera mitad de siglo son mucho mayores, debido a que hay muchas menos películas y esto hace que cada película tenga mucho más peso.
Tamién me parece necesario considerar que el total (100%) no se corresponderá al total de películas, sino al total de generos. Si, por ejemplo, a una película le corresponden 3 géneros, estará representada 3 veces.

De esta gráfica podemos decir que todos los géneros tienen una distribución parecida con el tiempo, aumentando ligeramente en los 80 y mucho más en los 90.
